# **CALCULADORA DE PORCENTAJES**

```python
# CALCULAR PORCENTAJE DE CRECIMIENTO
valor_inicial = float(input('Valor inicial: '))
valor_actual = float(input('Valor actual: '))
crecimiento = ((valor_actual - valor_inicial) / valor_inicial) * 100
crecimiento_redondeado = round(crecimiento,2)
print('El crecimiento ha sido ' + str(crecimiento_redondeado) + '%')
```

```python
# CALCULAR PORCENTAJE DE CONSECUCIÃ“N DE OBJETIVOS
objetivo = float(input('Objetivo: '))
resultado = float(input('Resultado: '))
objetivo_conseguido = ((resultado * 100) / objetivo)
objetivo_redondeado = round(objetivo_conseguido,2)
print('Hemos conseguido un ' + str(objetivo_redondeado) + '% del objetivo')
```
